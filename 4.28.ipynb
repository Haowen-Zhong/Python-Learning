```python
'''
异常:
情况1:出现多种异常
所以可以这么来：
try:
    xxxxxx
except 异常类型1:
    print.....
except 类型2:
    print....
except Exception(这是可能出现我们没有预想到的问题，就用这个祖宗辈的错误类型):
finally
'''
def func(a,b):
    try:
        return a/b
    except Exception:
        #如果把Exception放在最前面的话由于它是祖宗类如果碰到一些可以预计的错误直接被接受了
        #因此若想用Exception我们应该要放在最下面来处理我们无法预计的问题。
        return "emmm"
a = func(1,0)
print(a)
```

```python
#如果想偷懒还有牛逼方法！
#真的牛逼，直接告诉我们出错的原因。那直接用这个不就好了！！！！
def func(a,b):
    try:
        c = a/b
        return c
    except Exception as err:
        #如果把Exception放在最前面的话由于它是祖宗类如果碰到一些可以预计的错误直接被接受了
        #因此若想用Exception我们应该要放在最下面来处理我们无法预计的问题。
        return f"emmm,{err}"
a = func(1,0)
print(a)
```

```python
def func():
    stream = None
    try:
        #这里声明的只属于try这一小模块
        stream = open(r"c:\p1\aa.txt")
        container = stream.read()
        print(container)
        return 1
    except Exception as err:
        print(err)
        return 2
    finally:
        print('--------finally-------')
        if stream:
            stream.close()
        return 3
x = func()
print(x)
```

# 列表推导式 还有字典推导式 集合推导式

```python
#旧的列表 ----> 新的列表
#1.列表推导式： 格式：[表达式 for 变量 in 旧的列表] or [表达式 for 变量 in 旧的列表 if 条件]
#过滤掉长度小于或者等于3的任命
#这个是非常有用的！！！
names = ['tom','lily','abc','jack','steven','bob']
new_name = [name.capitalize() for name in names if len(name)>3]
print(new_name)
new_name = [name.upper() for name in names if len(name)>3]
print(new_name)
```

```python
#将1到100之间能被3整除的组成一个列表
number = [number for number in range(1,101) if number%3==0]
print(number)
```

```python
#0~5的偶数，0~10的奇数 eg.[(0,1),(0,3),(0,5),(0,7)...]
#我这么写的语法居然可以......我都懵了
import time
time1 = time.time()
number = [(i,j) for i in range(0,50) for j in range(0,50) if i%2==0 if j%2!=0]
time2 = time.time()
delta1 = time2 - time1
#print(number)
#如果老老实实用循环：
def number():
    list1 = []
    for i in range(0,50):
        for j in range(0,50):
            if i%2==0 and j%2!=0:
                list1.append((i,j))
    return list1
time1 = time.time()
list1 = number()
time2 = time.time()
delta2 = time2 - time1
#print(list1)
print(f"列表推导式所用时间{delta1},循环所用时间{delta2}")
#时间有波动。
```
