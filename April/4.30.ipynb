```python
def task1(n):
    for i in range(n):
        print(f"æ­£åœ¨æ¬ç¬¬{i+1}å—ğŸ§±ï¼")
        yield None#å°±åˆ©ç”¨yieldçš„æš‚åœä»»åŠ¡
def task2(n):
    for i in range(n):
        print(f"æ­£åœ¨å¬ç¬¬{i+1}é¦–æ­Œ!")
        yield None#å°±åˆ©ç”¨yieldçš„æš‚åœä»»åŠ¡
g1 = task1(5)
g2 = task2(3)
while True:
    try:
        g1.__next__()#å…ˆåšä½ 
        g2.__next__()#å†åšä»–
    except:
        break
```
